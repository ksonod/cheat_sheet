# Preparation

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import gridspec
```

```python
# preparation of data for visualization
np.random.seed(0)
x1 = np.linspace(0,10,101)

perturbation = np.zeros_like(x1)
perturbation[40:50] = 0.5 * np.sin(18*x1[40:50]) + 0.3*np.random.randn(x1[40:50].shape[0])

y1 = np.exp(x1*0.14) + 0.1*np.random.randn(x1.shape[0]) -3
y2 = np.cos(x1) + 0.1*np.random.randn(x1.shape[0]) + perturbation 
```

# Data Visualization


With matplotlib, you can draw graphs in two different ways:
- Matlab style
- Object-oriented style  

https://matplotlib.org/matplotblog/posts/pyplot-vs-object-oriented-interface/

## Matlab style

### Plot

```python
plt.rcParams.update({"font.size": 14})
plt.figure(figsize=(12,6))
plt.plot(x1, y1, label="sample A", color = "black", marker="o", markersize=9)
plt.plot(x1,y2, label="sample B", color = "red", marker="x", markersize=9)
plt.legend(loc="upper left", frameon=False)
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))  # x axis contains only integers 
plt.gca().get_yaxis().set_major_locator(ticker.MaxNLocator(integer=True))  # y axis contains only integers 
plt.grid()
plt.xlim([0,10])
plt.ylim([-2,2])
plt.annotate(text="Perturbation", xy=(4,0.9))
plt.xlabel("Time [/ 10$^{-9}$ s]")
plt.ylabel("Signal intensity [arb. units]")
plt.show()
```

### Subplot

```python
plt.rcParams.update({"font.size": 14})
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(x1, y1, label="sample A", color = "black", marker="o", markersize=9)
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))  # x axis contains only integers 
plt.gca().get_yaxis().set_major_locator(ticker.MaxNLocator(integer=True))  # y axis contains only integers 
plt.xlim([0,10])
plt.ylim([-2,2])
plt.xlabel("Time [/ 10$^{-9}$ s]")
plt.ylabel("Signal intensity [arb. units]")
plt.title("signal A")

plt.subplot(122)
plt.plot(x1,y2, label="sample B", color = "red", marker="x", markersize=9)
plt.xlim([0,10])
plt.ylim([-2,2])
plt.title("signal B")
plt.xlabel("Time [/ 10$^{-9}$ s]")
plt.ylabel("Signal intensity [arb. units]")

plt.show()
```

## Object-oriented style

### Basic plot 1

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
ax.plot(x1, y1, label="sample A", color = "black", marker="o", markersize=9)
ax.plot(x1,y2, label="sample B", color = "red", marker="x", markersize=9)
ax.legend(loc="upper left", frameon=False)
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))  # x axis contains only integers
ax.yaxis.set_major_locator(ticker.MaxNLocator(integer=True))  # y axis contains only integers
ax.grid()
ax.set_xlim([0,10])
ax.set_ylim([-2,2])
ax.annotate(text="Perturbation", xy=(4,0.9))
ax.set_xlabel("Time [/ 10$^{-9}$ s]")
ax.set_ylabel("Signal intensity [arb. units]")
None
```

### Basic plot 2

Alternatively, you can use  
fig, ax = plt.subplots()

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x1, y1, label="sample A", color = "black", marker="o", markersize=9)
ax.plot(x1,y2, label="sample B", color = "red", marker="x", markersize=9)
ax.legend(loc="upper left", frameon=False)
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))  # x axis contains only integers
ax.yaxis.set_major_locator(ticker.MaxNLocator(integer=True))  # y axis contains only integers
ax.grid()
ax.set_xlim([0,10])
ax.set_ylim([-2,2])
ax.annotate(text="Perturbation", xy=(4,0.9))
ax.set_xlabel("Time [/ 10$^{-9}$ s]")
ax.set_ylabel("Signal intensity [arb. units]")
None
```

### Subplot

https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

#### Subplot 1

```python
fig = plt.figure(figsize=(14,4))
ax0 = fig.add_subplot(121)
ax0.plot(x1, y2, color = "red", marker="x", markersize=9)

ax1 = fig.add_subplot(122)
ax1.plot(x1, y1, color = "black", marker="o", markersize=9)

None
```

#### Subplot 2

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14,4))
ax[0].plot(x1, y2, color = "red", marker="x", markersize=9)
ax[1].plot(x1, y1, color = "black", marker="o", markersize=9)
None
```

#### Subplot with different sizes

```python
fig = plt.figure(figsize=(14,4))
spec = gridspec.GridSpec(nrows=1, ncols=2, width_ratios=[3, 1])

ax0 = fig.add_subplot(spec[0])
ax0.plot(x1, y2, color = "red", marker="x", markersize=9)
ax1 = fig.add_subplot(spec[1])
ax1.plot(x1, y1, color = "black", marker="o", markersize=9)
None
```

#### Subplot without spaces

```python
fig = plt.figure(figsize=(14,7))
gs = fig.add_gridspec(3, hspace=0)
axs = gs.subplots(sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x1, y2, 'ko-')
axs[1].plot(x1, y1, 'ko-')
axs[2].plot(x1, y1-y2, 'ko-')

for ax in axs:
    ax.label_outer()  
```
