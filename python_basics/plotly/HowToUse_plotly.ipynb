# plotly

```python
import plotly.offline
import plotly.graph_objs as go
import numpy as np

ydat1 = np.random.rand(100)+1 # sample data
ydat2 = np.random.rand(100)+1 # sample data


data = [
    go.Scatter(y=ydat1, name='data 1'),
    go.Scatter(y=ydat2, name='data 2')
]

layout = go.Layout(
    xaxis={"title":"X"},
    yaxis={"title":"Y"},
    width=650, height=450,
    plot_bgcolor='rgba(0,0,0,0)',
    legend=dict(x=0.02, y=1)
)
```

```python
fig = go.Figure(data=data, layout = layout)

fig.update_xaxes(ticks="inside", tickwidth=2, tickcolor='black',linecolor='black', ticklen=4,mirror=True)
fig.update_yaxes(ticks="inside", tickwidth=2, tickcolor='black',linecolor='black', ticklen=4,mirror=True, range=[0,2.3])
fig.update_layout(legend_orientation="h")
fig.show()
```
