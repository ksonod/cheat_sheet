# Abstract

In this notebook, an image in a folder is loaded into a code using the following libraries:
- matplotlib
- pillow
- scikit image
- Open CV

# Libraries

```python
import os
import requests
import matplotlib.pyplot as plt
import numpy as np
```

# Preparation of an image

```python
name='picture.jpg'
pic = requests.get('https://picsum.photos/500/500/?random') # random images

with open(name,'wb') as pic_file:
    pic_file.write(pic.content)
    
datpath=os.path.join( os.getcwd() , 'picture.jpg')
```

# Various methods for loading an image

## [matplotlib.pyplot.imread](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imread.html?highlight=imread#matplotlib.pyplot.imread)

```python
img_array= plt.imread(datpath) # numpy array
# img_array= plt.imread(datpath, 0) # png

print('data type:',type(img_array))
print('shape:',img_array.shape)

plt.imshow(img_array)
plt.show()
```

## [Pillow](https://pillow.readthedocs.io/en/stable/)

### Color

```python
from PIL import Image

img=Image.open(datpath)
img
```

Let us check the data type.

```python
type(img)
```

If you add ".size," you can check the size of an image.

```python
print("width: {}, height: {}".format(img.size[0], img.size[1]))
```

The image can be converted into a numpy array.

```python
img_array=np.array(img)
print('data type:',type(img_array))
print('shape:',img_array.shape)
```

```python
plt.imshow(img_array)
plt.show()
```

### Gray scale

```python
img=Image.open(datpath).convert('L')
img
```

```python
img_array=np.array(img)
print('data type:',type(img_array))
print('shape:',img_array.shape)

plt.imshow(img_array, cmap='gray')
plt.show()
```

## [Scikit image](https://scikit-image.org)

### Color

```python
from skimage import io

img_array=io.imread(datpath)

print('data type:',type(img_array))
print('shape:',img_array.shape)

plt.imshow(img_array)
plt.show()
```

### Gray scale

```python
img_array=io.imread(datpath, as_gray=True)

print('data type:',type(img_array))
print('shape:',img_array.shape)

plt.imshow(img_array,cmap='gray')
plt.show()
```

## OpenCV

### Color

```python
import cv2

img_array=cv2.imread(datpath)

print('data type:',type(img_array))
print('shape:',img_array.shape)

plt.imshow(img_array)
plt.show()
```

The order of the color is BGR in OpenCV. However, the matplotlib adopts RGB order. Therefore, it is necessary to convert from BGR to RGB.

```python
plt.imshow( cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB) )
```

If you run the next cell, a new window will show up.

```python
cv2.imshow(datpath,img_array)
cv2.waitKey(0) # waits until a key is pressed
cv2.destroyAllWindows()
```

### Gray scale

```python
img_array=cv2.imread(datpath, cv2.IMREAD_GRAYSCALE)
#img_array=cv2.imread(datpath, 0) 


print('data type:',type(img_array))
print('shape:',img_array.shape)

plt.imshow(img_array, cmap='gray')
plt.show()
```
